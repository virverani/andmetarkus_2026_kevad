## if/else

```python
def make_sense_of_this_year(year: int):
    if year < 0:
        print("B.C")
        # year = -90
        if year <= -10000:
            print("Stone age...")
    elif year == 0:
        print("Happy birthday, Jesus!")
        
    elif year == 1978:
        print("Happy birthday, Mati!")
    else:
        print("A.D.")

make_sense_of_this_year(-90)
make_sense_of_this_year(year=-90)
```

```python
def format_year(year: int) -> str:
    if year < 0:
        return f"{abs(year)} B.C"
    return f"{year} A.D."

print(format_year(-90))
print(format_year(2025))
```

```python
from sklearn.datasets import load_iris
import pandas as pd

iris_data = load_iris()
iris = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
iris['target'] = iris_data.target
iris['species'] = iris['target'].map(dict(enumerate(iris_data.target_names)))
iris.pop('target')
iris = iris.rename(columns={
    "sepal length (cm)": "sepal_length",
    "sepal width (cm)": "sepal_width",
    "petal length (cm)": "petal_length",
    "petal width (cm)": "petal_width"
})
iris
```

```python
iris.to_csv("iris.csv", index=False)
```

```python
duckdb.sql("""
    SELECT
        species,
        avg(petal_length) AS avg_petal_length
    FROM iris
    GROUP BY species
    ORDER BY avg_petal_length DESC
""").df()
```

- vaja leiad kogu hind
- kui hind üle 100, siis 20% alla
- kui hind üle 50, siis 10% alla

sammud:
- leida allahindlusega hind igale hinnale
- kokku liita

```python

def get_discount(price: int) -> float:
    """
    See funktsioon arvutab allahindluseid.
    """
    if price > 100:
        return price * 0.8
    if price > 50:
        return price * 0.9
    return price

def get_price(kalkun_on_tore_loom: list[int]) -> float:
    tulemus = 0
    for hind in kalkun_on_tore_loom:
        allahindlusega_hind = get_discount(hind)
        tulemus = tulemus + allahindlusega_hind
    return tulemus

def get_price_short(kalkun_on_tore_loom: list[int]) -> float:
    return sum([get_discount(hind) for hind in kalkun_on_tore_loom])


hinnad = [120, 45, 89, 200, 33, 75, 150, 10, 99, 101]
get_price_short(hinnad)
```
